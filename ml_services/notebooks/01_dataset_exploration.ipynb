# TrustNewsAI Dataset Exploration

Goal: understand the Kaggle Fake and Real News Dataset before training any model.

Expected raw files:

- `../data/raw/True.csv`
- `../data/raw/Fake.csv`

````python
from pathlib import Path

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

RAW_DIR = Path('../data/raw')
PROCESSED_DIR = Path('../data/processed')

true_path = RAW_DIR / 'True.csv'
fake_path = RAW_DIR / 'Fake.csv'

true_path.exists(), fake_path.exists()
````

## Load Raw Data

This cell will fail until you download the Kaggle dataset and place the files in `data/raw/`.

````python
real_df = pd.read_csv(true_path)
fake_df = pd.read_csv(fake_path)

real_df['label'] = 'REAL'
fake_df['label'] = 'FAKE'

raw_df = pd.concat([real_df, fake_df], ignore_index=True)
raw_df.head()
````

````python
raw_df.info()
raw_df['label'].value_counts()
````

## Basic Quality Checks

````python
raw_df.isna().sum()
````

````python
raw_df.duplicated(subset=['title', 'text']).sum()
````

````python
raw_df['content'] = raw_df['title'].fillna('') + ' ' + raw_df['text'].fillna('')
raw_df['content_length'] = raw_df['content'].str.len()

raw_df.groupby('label')['content_length'].describe()
````

````python
sns.countplot(data=raw_df, x='label')
plt.title('Label Balance')
plt.show()
````
